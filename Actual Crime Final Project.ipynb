#Subsetting data frame for Violent crimes 

```python
subV = df[ ['Geo_FIPS', 'Geo_NAME',
       'Geo_COUNTY', 'SE_T005_001', 'SE_T005_002',
       'SE_T005_003', 'SE_T005_005', 'SE_T005_006'] ]

subV.head(3)
```

```python
subV.columns
```

T005_002:      Murders and Nonnegligent Manslaughters Rate
        T005_003:      Rape Revised Definition Rate
        T005_004:      Rape Legacy Definition Rate
        T005_005:      Robberies Rate
        T005_006:      Aggravated Assaults Rate

##Melting data frame and renaming columns for violent crimes 

```python
subV.rename(columns={"SE_T005_001": "Total", "SE_T005_002": "Manslaughter", "SE_T005_003": "Rape", "SE_T005_005": "Robberies", "SE_T005_006": "Assualt"}, inplace=True)

melted = pd.melt(subV, id_vars=['Geo_FIPS', 'Geo_NAME', 'Geo_COUNTY'], value_vars=['Total','Manslaughter', 'Rape', 'Robberies', 'Assualt'], var_name='Type', value_name="Rate")
melted = melted[melted['Rate'].notna()]
melted
```

##Pie Charts for Violent Crimes 2019

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.0f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


temp = melted[ melted.Geo_NAME == "New York"] 
data = temp[1:]['Rate'].values
lbls = temp[1:]['Type'].values


plt.gcf().set_size_inches(6, 6)
plt.pie(data, labels=lbls, autopct='%.1f%%')
plt.show()
```

```python
temp = melted[ melted.Geo_NAME == "California"] 
data = temp[1:]['Rate'].values
lbls = temp[1:]['Type'].values


plt.gcf().set_size_inches(6, 6)
plt.pie(data, labels=lbls, autopct='%.1f%%')
plt.show()
```
